```python
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import pandas as pd


file_path = os.path.join(path, 'ecommerce_customer_churn_dataset.csv')
df = pd.read_csv(file_path)
```

```python
#. What is the overall customer churn rate?
churn_rate = df_clean['Churned'].value_counts(normalize=True) * 100
churn_rate
```

```python
labels = ['Retained Customers', 'Churned Customers']
sizes = [churn_rate[0], churn_rate[1]]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,
    explode=(0, 0.05))  # slightly separate churned slice
plt.title('Overall Customer Churn Rate')
plt.tight_layout()
plt.show()
```

```python
behavior_cols = ['Login_Frequency','Session_Duration_Avg','Pages_Per_Session','Cart_Abandonment_Rate','Days_Since_Last_Purchase']

behavior_summary = (df_clean.groupby('Churned')[behavior_cols].mean().rename(index={0: 'Retained', 1: 'Churned'}))

behavior_summary
```

```python
#. Which behavioral factors most strongly influence churn?
behavior_summary.T.plot(
    kind='bar',
    figsize=(10, 6)
)

plt.title('Behavioral Factors Influencing Customer Churn')
plt.ylabel('Average Value')
plt.xlabel('')
plt.xticks(rotation=45)
plt.legend(title='Customer Status')
plt.tight_layout()
plt.grid(False) 
plt.show()
```

```python
# Define High LTV threshold (75th percentile)
ltv_threshold = df_clean['Lifetime_Value'].quantile(0.75)

df_clean['LTV_Segment'] = df_clean['Lifetime_Value'].apply(
    lambda x: 'High LTV' if x >= ltv_threshold else 'Low / Medium LTV'
)

df_clean['LTV_Segment'].value_counts()
```

```python
ltv_features = [
    'Total_Purchases',
    'Average_Order_Value',
    'Login_Frequency',
    'Session_Duration_Avg',
    'Pages_Per_Session',
    'Cart_Abandonment_Rate',
    'Days_Since_Last_Purchase',
    'Discount_Usage_Rate'
]

ltv_summary = df_clean.groupby('LTV_Segment')[ltv_features].mean()

ltv_summary
```

```python
#. What characteristics define high-lifetime-value customers?
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

metrics = [
    'Total_Purchases',
    'Average_Order_Value',
    'Login_Frequency',
    'Cart_Abandonment_Rate'
]

for i, metric in enumerate(metrics):
    sns.boxplot(
        data=df_clean,
        x='LTV_Segment',
        y=metric,
        ax=ax[i//2, i%2],
        palette='Set3')
    ax[i//2, i%2].set_title(metric)
    ax[i//2, i%2].grid(False)

plt.suptitle('Distribution of Key Behaviors by Lifetime Value Segment')
plt.tight_layout()
plt.show()
```

```python
signup_churn = (
    df_clean
    .groupby('Signup_Quarter')['Churned']
    .mean()
    .mul(100)
    .reset_index(name='Churn_Rate')
)
signup_churn
```

```python
signup_ltv = (
    df_clean
    .groupby('Signup_Quarter')['Lifetime_Value']
    .mean()
    .reset_index(name='Avg_Lifetime_Value')
)
signup_ltv
```

```python
numeric_cols = df_clean.select_dtypes(include='number')

churn_corr = (
    numeric_cols
    .corr()['Churned']
    .sort_values()
)

churn_corr.head(10), churn_corr.tail(10)
```

```python
#. What are the Top churn predictors
top_predictors = churn_corr.drop('Churned').sort_values().head(10)

plt.figure(figsize=(8, 6))
sns.barplot(
    x=top_predictors.values,
    y=top_predictors.index,
    palette='coolwarm'
)

plt.title('Top Behavioral Predictors of Customer Churn')
plt.xlabel('Correlation with Churn')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Create engagement score (you can adjust features if needed)
df_clean['Engagement_Score'] = (
    df_clean['Login_Frequency'] +
    df_clean['Session_Duration_Avg'] +
    df_clean['Pages_Per_Session']
)

# Create quantile-based segments
df_clean['Engagement_Level'] = pd.qcut(
    df_clean['Engagement_Score'], 
    q=3, 
    labels=['Low Engagement', 'Medium Engagement', 'High Engagement']
)

df_clean['LTV_Level'] = pd.qcut(
    df_clean['Lifetime_Value'], 
    q=3, 
    labels=['Low LTV', 'Medium LTV', 'High LTV']
)
```

```python
heatmap_data = (
    df_clean
    .groupby(['Engagement_Level', 'LTV_Level'])['Churned']
    .mean()
    .unstack()
)
```

```python
#. Which customer segments should be prioritized for retention strategies?
plt.figure(figsize=(9, 6))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".1%",
    cmap="Reds",
    linewidths=0.5,
    cbar_kws={'label': 'Churn Rate'}
)

plt.title('Customer Churn by Engagement and Lifetime Value')
plt.xlabel('Lifetime Value Segment')
plt.ylabel('Engagement Level')

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
